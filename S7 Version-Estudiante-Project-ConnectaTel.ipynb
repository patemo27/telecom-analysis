## 🧩Paso 3: Limpieza básica de datos

### 3.1 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Aplicar reglas de limpieza para reemplazar valores sentinels y corregir fechas imposibles.

**Instrucciones:**  
- En `age`, reemplaza el sentinel **-999** con la mediana.
- En `city`, reemplaza el sentinel `"?"` por valores nulos (`pd.NA`).  
- Marca como nulas (`pd.NA`) las fechas fuera de rango.

```python
# Reemplazar -999 por la mediana de age
age_mediana = users.loc[users['age'] != -999, 'age'].median()
users['age'] = users['age'].replace(-999, age_mediana)

# Verificar cambios
users['age'].describe()
```

```python
# Reemplazar ? por NA en city
users['city'] = users['city'].replace("?", pd.NA)

# Verificar cambios
print("Valores únicos actualizados en 'city':")
print(users['city'].unique())
print("\nConteo de nulos en la columna:")
print(users['city'].isna().sum())
```

```python
# Marcar fechas futuras como NA para reg_date
users.loc[users['reg_date'].dt.year > 2024, 'reg_date'] = pd.NaT

# Verificar cambios
print("Años presentes después de la limpieza:")
print(users['reg_date'].dt.year.unique())

print("\nCantidad de fechas futuras convertidas a nulas:")
print(users['reg_date'].isna().sum())
```

### 3.2 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Decidir qué hacer con los valores nulos según su proporción y relevancia.

**Instrucciones:**
- Verifica si los nulos en `duration` y `length` son **MAR**(Missing At Random) revisando si dependen de la columna `type`.  
  Si confirmas que son MAR, **déjalos como nulos** y justifica la decisión.

```python
# Verificación MAR en usage (Missing At Random) para duration
verificacion_duration = usage.groupby('type')['duration'].apply(lambda x: x.isna().value_counts(dropna=False)).unstack(fill_value=0)
verificacion_duration.columns = ['Tiene Dato (Valid)', 'Es Nulo (NaN)']
print("=============================================================")
print("VERIFICACIÓN DE NULOS EN DURATION SEGÚN EL TIPO DE USO")
print("=============================================================")
print(verificacion_duration)
```

```python

# Verificación MAR en usage (Missing At Random) para length
verificacion_length = usage.groupby('type')['length'].apply(lambda x: x.isna().value_counts(dropna=False)).unstack(fill_value=0)
verificacion_length.columns = ['Tiene Dato (Valid)', 'Es Nulo (NaN)']
print("=============================================================")
print("VERIFICACIÓN DE NULOS EN LENGTH SEGÚN EL TIPO DE USO")
print("=============================================================")
print(verificacion_length)
```

Los resultados confirman que los valores nulos en duration y length son MAR, ya que su ausencia depende por completo de la variable categórica type. Al observar las tablas, el 100% de los nulos en la duración (22,076) corresponden a registros de tipo text, mientras que la inmensa mayoría de los nulos en longitud pertenecen a registros de tipo call (exceptuando un margen mínimo de error de captura de 12 filas).

Se deben dejar como nulos ya que esta falta de datos no se debe a una pérdida accidental de información, sino a la lógica estructural del negocio: un mensaje de texto no posee duración en minutos, y una llamada telefónica no genera caracteres de texto. Intentar rellenar o imputar estos campos con la media o la mediana alteraría las propiedades estadísticas reales de cada servicio, por lo que conservar los nulos es la única estrategia metodológicamente correcta para no sesgar los análisis de consumo posteriores.

Esta variable es la que muestra la mayor cantidad de outliers, formando una densa línea de puntos atípicos que inicia desde los 62 minutos y se dispersa a lo largo del eje hasta llegar a picos extremos cercanos a los 160 minutos por llamada.

```python

# Calcular límites con el método IQR
columnas_limites = ['cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

print("=============================================================")
print("LÍMITES SUPERIORES DETECTADOS POR IQR")
print("=============================================================\n")

for col in columnas_limites:
    q1 = user_profile[col].quantile(0.25)
    q3 = user_profile[col].quantile(0.75)

    iqr = q3 - q1

    limite_superior = q3 + (1.5 * iqr)

    cantidad_outliers = (user_profile[col] > limite_superior).sum()
    
    print(f"Columna: {col}")
    print(f"  -> Q1: {q1:.2f} | Q3: {q3:.2f} | IQR: {iqr:.2f}")
    print(f"  -> Límite Superior: {limite_superior:.2f}")
    print(f"  -> Cantidad de outliers: {cantidad_outliers} usuarios\n")

```

```python
# Revisa los limites superiores y el max, para tomar la decisión de mantener los outliers o no
user_profile[columnas_limites].describe()
```

Una diferencia de poco más de 5 mensajes por encima del límite técnico es perfectamente viable. Tiene un grupo de 30 usuarios que realiza entre 11 y 15 llamadas no constituye una anomalía matemática o un error del sistema, sino que identifica ocn precisión el segmento de clientes con alta actividad telefónica, un dato de gran valor para estrategias de segmentación de planes comerciales.

Diseñar un plan intermedio capturaría a los usuarios del plan Básico que están al límite de su consumo pero que temen dar el salto económico al plan Premium.
- Lanzamiento de un Plan Corporativo / Ilimitado ("Connecta Pro"): Dado que los outliers (Heavy Users) consumen volúmenes masivos de datos y minutos, se les puede empaquetar en un nuevo plan de mayor costo que incluya priorización de red y herramientas de productividad, incrementando aún más el margen de ganancia con clientes que ya demostraron no tener problema en consumir de forma extrema.
- Estrategia de Migración Directa (Upselling): Identificar a los usuarios del plan Básico cuyo consumo de minutos o mensajes se encuentre cerca del "Límite Superior" calculado por el método IQR, y lanzarles campañas personalizadas de migración al plan Premium, ofreciéndoles un descuento en los primeros tres meses. Su propio comportamiento demuestra que el plan Básico ya les queda chico.

### Análisis ejecutivo

⚠️ **Problemas detectados en los datos**
- Presencia masiva de nulos por diseño: Se detectó que el 55.2% de los datos en la columna duration y el 44.8% en length eran nulos, causados por la lógica del negocio (los mensajes no tienen duración y las llamadas no tienen caracteres).
- Presencia de outliers: Las tres variables de consumo presentaron un fuerte sesgo a la derecha con usuarios que generan un tráfico atípico muy elevado, los cuales no debían ser borrados por representar comportamiento real de alto valor.


🔍 **Segmentos por Edad**
- Distribución Uniforme: La edad de los usuarios se distribuye de manera equitativa y homogénea en toda la muestra, sin presentar concentraciones atípicas en ningún rango.
- Indiferencia de Elección: No existe una correlación entre la edad y el tipo de servicio contratado; tanto jóvenes como adultos mayores adquieren los planes Básico y Premium en proporciones muy similares.


📊 **Segmentos por Nivel de Uso**
- Usuarios Básico: Este segmento muestra un comportamiento de consumo muy compacto y limitado, manteniéndose estrictamente en rangos bajos de minutos y mensajes para no exceder su tarifa.
- Usuarios Premium: Este segmento aprovecha la libertad de su plan, expandiendo su consumo de forma masiva hacia la derecha de la distribución y concentrando los valores más valiosos y rentables para la empresa.


➡️ Esto sugiere que la estrategia de ConnectaTel no debe enfocarse en segmentaciones demográficas tradicionales (como la edad), sino en una segmentación por comportamiento y volumen de uso, ya que es el consumo real el que define el valor del cliente y la carga en la infraestructura de la red.


💡 **Recomendaciones**
- Lanzamiento de un Plan Flex: Crear un plan intermedio para capturar a los usuarios del plan Básico que se encuentran al límite de su consumo pero que perciben el plan Premium como un salto económico muy drástico.
- Campañas de Upselling Dirigidas: Identificar mediante el método IQR a los usuarios del plan Básico que registran consumos cercanos al límite superior y ofrecerles una migración directa al plan Premium con un descuento por hacer la transición.
